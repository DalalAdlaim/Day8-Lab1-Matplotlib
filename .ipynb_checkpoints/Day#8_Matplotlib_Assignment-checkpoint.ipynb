```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Q1: Create two charts using list and dataframe

```python
# Create a chart using lists, 
plt.plot([1,2,3,4,5]) 

df = pd.DataFrame({
    'name':['john','mary','peter'],
    'age':[23,78,22] 
})
df.plot(kind='bar',x='name',y='age')
```

## Q2: Create pie chart with equal explode values.

```python
plt.pie(np.arange(1,5), labels=['a','b','c', 'd'], explode = [.1,.1,.1,.1]);
```

## Q3: Create a scatter chart with circle marker, alpha = 0.4, and size = 50.

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y,alpha = 0.4,s = 50,marker = 'o')
```

## Q4: Add the following to your chart:
- title placed in the left and has a red color
- x and y label with font size = 16

```python
x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y,alpha = 0.4,s = 50,marker = 'o');
plt.title('scatter', loc='left', color='red');
plt.xlabel('X',fontsize=16)
plt.ylabel('Y',fontsize=16)
```

## Q5: Create three line charts with Legends at left and colors

```python
plt.plot([1,2,3]) 
plt.plot([4,5,6]) 
plt.plot([7,8,9]) 
plt.legend(['line1',"line2","line3"], shadow = True, loc = 0)
```

## Q6: Create subplots with three charts and figsize=20,10

```python
plt.figure(figsize=[20,10])

#first
plt.subplot(1,3,1) 
plt.plot([1,2,3]) 
plt.title('ch1')

#second
plt.subplot(1,3,2)
plt.plot([1,2,0,4,5])
plt.title('ch2');

#
plt.subplot(1,3,3)
plt.plot([5,10,15,20])
plt.title('ch2');
```
